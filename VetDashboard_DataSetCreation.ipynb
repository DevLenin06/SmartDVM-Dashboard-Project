## This is where I merged appointment information and assessment information, I deleted some columns that I didn't think were important, but we might need them.


```python
merged_dfapoint_dfasses = pd.merge(appointment_df, assessment_df, on=['appointment_id'], how='inner')
```

```python
merged_dfapoint_dfasses.columns
```

```python
merged_dfapoint_dfasses.drop(columns=['appointment_type', 'appointment_reason',
       'is_admitted', 'boarding_type', 'boarding_size', 'appointment_status',
       'cancelation_reason', 'appointment_date', 'appointment_checkout_date',
       'date_created', 'is_cancelled', 'in_progress', 'is_completed',
       'clinic_id', 'patient_id', 'doctor_assigned_id', 'payment_made',
       'boarding_status'], inplace=True)
```

# First five rows of data

```python
merged_dfapoint_dfasses.head()
```

```python
merged1_withdfdiagnosis = pd.merge(merged_dfapoint_dfasses, diagnosis_df, on=['assessment_id'], how='inner')
```

```python
merged1_withdfdiagnosis
```

```python
merged1_withdfdiagnosis.columns
```

```python
merged1_withdfdiagnosis.drop(columns=['date_added', 'is_final', 'diagnosis_state'], inplace=True)
```

```python
merged1_withdfdiagnosis.info()
```

```python
merged1_withdfobjective = pd.merge(merged1_withdfdiagnosis, objective_df, on=['appointment_id'], how='inner')
```

```python
merged1_withdfobjective.head()
```

```python
merged1_withdfvital = pd.merge(merged1_withdfobjective, vitalfindings_df, on=['objective_id'], how='inner')
```

```python
finaldf = merged1_withdfvital
```

```python
finaldf.columns
```

```python
finaldf['Weight_content'] = finaldf['content'].where(finaldf['title'].str.contains('Weight', case=False, na=False))
finaldf['CRT_content'] = finaldf['content'].where(finaldf['title'].str.contains('CRT', case=False, na=False))
finaldf['RT_content'] = finaldf['content'].where(finaldf['title'].str.contains('RT', case=False, na=False))
```

```python
finaldf
```

```python
finaldf.info()
```

```python
finaldf.drop(columns=['title','content', 'assessment_id', 'diagnosis_id', 'objective_id', 'vitalfindings_id'])
```

```python
finaldf_combined = finaldf.groupby('appointment_id').agg(lambda x:', '.join(x.dropna().astype(str))).reset_index()
```

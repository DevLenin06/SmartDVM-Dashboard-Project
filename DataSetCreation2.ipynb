```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('animal_dis_me.csv')
```

```python
df.head()
```

```python
import random
```

```python
new_disease = "Feline Panleukopenia"
new_symptoms = ['vomiting', 'diarrhea', 'lethargy', 'fever', 'dehydration', 'sudden_death']
```

```python
for symptom in new_symptoms:
    if symptom not in df.columns:
        df[symptom] = 0.0
```

```python
print("New columns after initialization:", df.columns)
```

```python
df[new_symptoms] = df[new_symptoms].fillna(0)
```

```python
df.head()
```

```python
df[df['prognosis'] == "Feline Panleukopenia"]
```

```python
new_data = []
for _ in range(50):
    patient_data = {symptom: random.choice([0, 1]) for symptom in new_symptoms}

    patient_data['prognosis'] = new_disease
    for symptom in df.columns:
        if symptom not in new_symptoms and symptom != 'prognosis':
            patient_data[symptom] = random.choice([0, 1])

    new_data.append(patient_data)
```

```python
new_df = pd.DataFrame(new_data)
```

```python
df = pd.concat([df, new_df], ignore_index=True)
```
